# Time Series Analysis

For the time series analysis, we are interested in looking at removal of toxic chemicals trend over the past years. So, we will be using 'ChemicalDateRemoved'
column next. First, we will convert the date column to datetime format as you can see below.

```python
# Converting date column to datetime format
data['ChemicalDateRemoved'] = pd.to_datetime(data['ChemicalDateRemoved'])
```

```python
# Filtering out rows where 'ChemicalDateRemoved' is NaT (not a date)
chem_removed = data[data['ChemicalDateRemoved'].notna()]
chem_removed.head()
```

```python
# Extracting the year from 'ChemicalDateRemoved' and counting the removals per year
chem_removed['YearRemoved'] = chem_removed['ChemicalDateRemoved'].dt.year
chemicals_removed_per_year = chem_removed['ChemicalDateRemoved'].value_counts().sort_index()

# Plotting the trend of chemical removal over the years
plt.figure(figsize=(12, 8))
plt.plot(chemicals_removed_per_year.index, chemicals_removed_per_year.values, color='red', marker = 'o' )

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Number of Chemicals Removed')
plt.title('Trend of Chemical Removal Over the Years')

# Displaying the plot
plt.show()
```

This time series graph shows the trend of chemical removals from cosmetic and skincare products over the years. The x-axis represents the years from 2000 to 2020, while the y-axis reflects the number of chemicals removed each year. there is a sharp spike around 2010-2012, which likely indicates increased regulatory action or awareness regarding the safety of certain chemicals used in cosmetics. After this peak, there is still considerable activity in the removal of chemicals, though at lower levels.

# K-Means Clustering

K-Means Clustering, is a good choice for this dataset because it can help group chemicals based on their usage patterns, frequency, and other features such as the company or product category. This is very beneficial especially for finding out which chemicals are always together, those that are of the same product categories, or that are inserted or removed at the same time. Therefore, I selected K-means clustering as a model for my analysis.

```python
# Importing the necessary libraries for K-means clustering analysis
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Converting dates to numeric

data['InitialDateReported'] = pd.to_datetime(data['InitialDateReported'])
data['ChemicalDateRemoved'] = pd.to_datetime(data['ChemicalDateRemoved'])
```

```python
# Converting date columns into numeric
data['DaysSinceReported'] = (pd.Timestamp.today() - data['InitialDateReported']).dt.days
```

```python
# Handling categorical features by encoding, PrimaryCategoryId and SubCategory columns are encoded into numeric values using LabelEncoder
label_encoder = LabelEncoder()
data['PrimaryCategoryId'] = label_encoder.fit_transform(data['PrimaryCategoryId'])
data['SubCategory'] = label_encoder.fit_transform(data['SubCategory'])
```

```python
# Selecting relevant features for clustering

features = data[['ChemicalCount', 'DaysSinceReported', 'PrimaryCategoryId', 'SubCategory']]
```

```python
print(features.isnull().sum()) # Validating no missing values are present in these selected features
```

```python
# Standardizing the features to bring them to the same scale

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# Performing K-Means Clustering

kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Visualizing the clusters using the first and third features

plt.figure(figsize=(10, 6))
plt.scatter(scaled_features[:, 0], scaled_features[:, 3], c=data['Cluster'], cmap='viridis')
plt.title('K-Means Clustering of Chemicals')
plt.xlabel('Chemical Count (Scaled)')
plt.ylabel('Product SubCategory (Scaled)')
plt.show()
```

The K-Means clustering visualization here shows the relationship between Product Subcategory (y-axis) and Chemical Count (x-axis). The purple cluster typically involves chemicals with higher chemical counts. In fact, it is seen that the products with high chemical counts are assorted in the different subcategories as well. In the same way, the yellow cluster is situated at a smaller proportion below the chemical count chart. Furthermore, the teal cluster, which represents the lowest chemical count, is widely spread in the different cosmetic products subcategories.
Moreover, items with high chemical counts, for instance those that are situated on the far right side, are forming certain clusters where they are more likely to be included in others than in other categories.
Overall, purple cluster seems to be the most dominating product line with a larger number of chemical count products, whereas both yellow and teal clusters are linked with lower chemical count products, suggesting a possible difference in product types according to the level of chemical use. 


# Further Analysis - Predicting Future Trends with ARIMA model (AutoRegressive Integrated Moving Average)

```python
# Importing required libraries

from statsmodels.tsa.arima.model import ARIMA
from pandas.plotting import register_matplotlib_converters
import warnings
```

```python
# Suppressing warnings for cleaner output
warnings.filterwarnings("ignore")
register_matplotlib_converters()
```

```python
# Converting InitialDateReported to datetime
data['InitialDateReported'] = pd.to_datetime(data['InitialDateReported'])
```

```python
# Setting InitialDateReported as the index for time series analysis
data.set_index('InitialDateReported', inplace=True)

# Sorting by date to ensure chronological order
data = data.sort_index()
```

```python
# Plotting the historical data
plt.figure(figsize=(10,6))
plt.plot(data['ChemicalCount'], label='ChemicalCount')
plt.title('Chemical Count Over Time')
plt.xlabel('Date')
plt.ylabel('Chemical Count')
plt.legend()
plt.show()
```

```python
# Splitting data into train and test to prepare for our ARIMA model

train_size = int(len(data) * 0.8)  # 80% training data, 20% testing data
train, test = data['ChemicalCount'][:train_size], data['ChemicalCount'][train_size:]

# Validating the split
print(f'Training set length: {len(train)}, Test set length: {len(test)}')
```

```python
# Fitting ARIMA model on the training data
#p: Number of lag observations (autocorrelation)
#d: Degree of differencing (trend)
#q: Size of the moving average window

model = ARIMA(train, order=(2,0,2))  
model_fit = model.fit()

# Printing model summary
print(model_fit.summary())
```

```python
# Forecasting the future values based on test data length
forecast = model_fit.forecast(steps=len(test))

# Plotting the forecast against the actual test data
plt.figure(figsize=(10,6))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Test Data', color='blue')
plt.plot(test.index, forecast, label='Forecast', color='red')
plt.title('ARIMA Forecast vs Actual Chemical Count')
plt.xlabel('Date')
plt.ylabel('Chemical Count')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Calculating MSE and MAE
mse = mean_squared_error(test, forecast)
mae = mean_absolute_error(test, forecast)

print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
```

```python
# Forecasting future values (e.g., next 5 years)
forecast_steps = 10  # Number of future periods to forecast

future_forecast = model_fit.forecast(steps=forecast_steps)

# Plotting the future forecast (next 10 years)
plt.figure(figsize=(10,6))
plt.plot(data.index, data['ChemicalCount'], label='Historical Data')
last_date = data.index[-1]
future_dates = pd.date_range(last_date, periods=forecast_steps, freq='Y') 
plt.plot(future_dates, future_forecast, label='Future Forecast', color='red')
plt.title('Future Forecast of Chemical Count')
plt.xlabel('Year')
plt.ylabel('Chemical Count')
plt.legend()
plt.show()
```

In the graph above, the future forecast (red line) is predicting a flat trend in the chemical count for next 10 years. In fact, the red colour on the line graph is representing future predictions and is anticipating no growth in the count of chemicals for the next 10 years, which hardly seems to be accurate. ARIMA model is not picking up sufficient patterns or trends in the data. Upon some further research on ARIMA model, I found that data itself may not have strong trends or patterns for the ARIMA model to learn from, especially if the ChemicalCount over time has been fairly constant or fluctuating randomly. So, it could give such results.
